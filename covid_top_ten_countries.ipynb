```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('covid19_confirmed_global.csv')
df.head(10)
```

```python
df.info()
```

```python
df.drop(columns ='Province/State', inplace = True)
```

```python
df.head()
```

```python
last_week_df = df[["Country/Region","3/24/22", "3/25/22", "3/26/22","3/27/22","3/28/22","3/29/22","3/30/22"]]
```

```python
print(last_week_df)
```

```python
top_ten = last_week_df.nlargest(n=10, columns=["3/24/22", "3/25/22", "3/26/22","3/27/22","3/28/22","3/29/22","3/30/22"])
```

```python
top_ten.plot(x="Country/Region", y=["3/24/22", "3/25/22", "3/26/22","3/27/22","3/28/22","3/29/22","3/30/22"], kind="bar")
```

```python
sns.barplot(x='Country/Region',y='3/30/22',data=top_ten)
```
